## INTRODUCTION

#### Backround & Public Health Significance

Obesity is one of the most pressing public health challenges facing the United States today. Defined as a body mass index (BMI) of 30 or higher, obesity is associated with increased risk of serious chronic conditions including type 2 diabetes, cardiovascular disease, hypertension, and certain cancers. According to the Centers for Disease Control and Prevention (CDC), more than 40% of American adults currently live with obesity, placing a significant burden on both individual health and the broader healthcare system. Understanding the social and economic factors that drive obesity rates is essential for developing effective public health interventions and reducing this burden across the population.

#### What We Know & Knowledge Gaps

Existing research has consistently identified socioeconomic status as an important determinant of health outcomes, including obesity. Studies have shown that individuals with lower incomes face greater exposure to obesogenic environments including limited access to healthy foods, fewer opportunities for physical activity, and higher levels of chronic stress all of which can contribute to weight gain over time. However, much of the existing literature relies on small regional samples or older data, and fewer studies have examined current national-level patterns using surveillance data collected directly by public health agencies. It remains important to regularly assess whether income-based disparities in obesity persist and to what degree they are reflected in contemporary population data.

## BACKGROUND

Literature Review: Income and Obesity in Adults
1. Overview of Obesity as a Public Health Issue

Obesity is a major public health concern globally, associated with increased risk of chronic diseases such as cardiovascular disease, type 2 diabetes, and certain cancers. According to the Centers for Disease Control and Prevention (CDC), obesity prevalence among U.S. adults has risen significantly over the past several decades, reflecting complex interactions between behavioral, environmental, and socioeconomic factors.

Public health research increasingly frames obesity not simply as an individual behavioral issue but as a socially patterned condition, strongly influenced by structural determinants such as income, education, and access to resources.

2. Socioeconomic Status (SES) and Health Theory
Social Determinants of Health Framework

The social determinants of health (SDOH) framework posits that health outcomes are shaped by the conditions in which people are born, live, and work (World Health Organization, 2008). Income is a central determinant because it influences:

Access to nutritious food
Healthcare availability
Safe environments for physical activity
Health literacy

Lower income populations often face structural barriers that promote obesity, including food insecurity and limited access to healthy foods.

Fundamental Cause Theory

Link and Phelan (1995) introduced Fundamental Cause Theory, arguing that socioeconomic status is a “fundamental cause” of disease because it embodies access to resources that can be used to avoid health risks. Even as risk factors change over time, SES remains a persistent predictor of health disparities—including obesity.

3. Key Empirical Studies
3.1 Income and Obesity Correlation
Ogden et al. (2017) found that obesity prevalence differs by income level in the U.S., with higher rates often observed among lower-income adults, particularly women.
Drewnowski & Specter (2004) demonstrated that energy dense, nutrient poor foods are often cheaper, contributing to higher obesity rates in low income populations.

These studies support the hypothesis that economic constraints influence dietary patterns, leading to increased obesity risk.

3.2 Food Environment and Access
Larson, Story, & Nelson (2009) reviewed research on food environments and found that low-income neighborhoods have fewer supermarkets and more fast-food outlets.
This phenomenon is often described as “food deserts”, where residents have limited access to affordable, healthy food.

Additionally, newer research introduces the concept of “food swamps”—areas saturated with unhealthy food options—which may be even more predictive of obesity than food deserts.

3.3 Behavioral and Psychological Factors
Sobal & Stunkard (1989) established early evidence of an inverse relationship between SES and obesity in developed countries.
Chronic stress associated with poverty has been linked to obesity through physiological pathways (e.g., cortisol regulation) (Adam & Epel, 2007).

This highlights that obesity is not only about food access but also stress, mental health, and coping mechanisms.

3.4 Gender and Racial Disparities
Research shows that the relationship between income and obesity varies by gender and race.
Wang & Beydoun (2007) found stronger inverse associations between SES and obesity among women than men.
Structural inequalities, including systemic racism, further compound disparities in obesity rates.
4. Conceptual Models Informing the Research
Energy Balance Model vs. Ecological Models

Traditional obesity research focused on the energy balance model (calories in vs. calories out). However, modern public health emphasizes ecological models, which consider:

Individual behavior
Social networks
Built environment
Policy-level influences

Your research aligns more with this ecological perspective, recognizing that income shapes multiple layers of influence.

Life Course Perspective

The life course perspective suggests that exposure to socioeconomic disadvantage over time accumulates, increasing obesity risk. Early-life poverty is especially predictive of adult obesity.

5. Gaps in the Literature

Despite extensive research, several gaps remain:

Causality vs. correlation: Many studies are cross-sectional
Intersectionality: Limited research integrating income, race, gender, and geography simultaneously
Policy evaluation: More evidence needed on which interventions effectively reduce SES-related obesity disparities

Your study contributes by helping clarify the relationship using your dataset, potentially adding context-specific or updated evidence.

## Methods

Data Source

This study used publicly available data from the CDC's Nutrition, Physical Activity, and Obesity — Behavioral Risk Factor Surveillance System (BRFSS) dataset, accessed through the CDC's open data portal (data.cdc.gov). The dataset contains state and national level estimates of obesity prevalence among U.S. adults aged 18 and older, collected annually across multiple years and stratified by demographic and socioeconomic characteristics including age, sex, race/ethnicity, education, and income level.

Study Population

The analysis focused on national level estimates of adult obesity prevalence stratified by income level. Income was categorized into groups as defined by the CDC: less than $15,000, $15,000–$24,999, $25,000–$34,999, $35,000–$49,999, and $50,000 or greater. Only respondents classified under the "Total" stratification category were included in the national trend and state-level analyses to ensure comparability across groups.

Data Analysis

Data were filtered to include only responses to the question: "Percent of adults aged 18 years and older who have obesity." Obesity prevalence values were averaged across available survey years for each income group. Three visualizations were produced: (1) a line chart showing national adult obesity prevalence over time, (2) a horizontal bar chart comparing obesity rates across U.S. states in the most recent available year, and (3) a horizontal bar chart comparing obesity prevalence across income groups at the national level. All analyses were conducted in Python using the pandas, matplotlib, and numpy libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Load data
url = "https://data.cdc.gov/api/views/hn4x-zwk7/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url, low_memory=False)

# Filter to obesity rates, overall population
obesity = df[
    (df['Topic'] == 'Obesity / Weight Status') &
    (df['Question'] == 'Percent of adults aged 18 years and older who have obesity') &
    (df['StratificationCategory1'] == 'Total') &
    (df['Data_Value_Type'] == 'Value')
].copy()

obesity['Data_Value'] = pd.to_numeric(obesity['Data_Value'], errors='coerce')
obesity = obesity.dropna(subset=['Data_Value'])

# --- Plot 1: Obesity rate over time with KDE ---
national = obesity[obesity['LocationDesc'] == 'National']
by_year = national.groupby('YearStart')['Data_Value'].mean().reset_index()

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 8))

# Line chart
ax1.plot(by_year['YearStart'], by_year['Data_Value'], marker='o', color='#e75480', linewidth=2)
ax1.fill_between(by_year['YearStart'], by_year['Data_Value'], alpha=0.15, color='#e75480')
ax1.set_title('US Adult Obesity Rate Over Time', fontsize=14)
ax1.set_xlabel('Year')
ax1.set_ylabel('Obesity Rate (%)')
ax1.set_ylim(0, 50)

# KDE of obesity values over all years
kde_data = by_year['Data_Value'].values
kde = gaussian_kde(kde_data)
x_range = np.linspace(kde_data.min() - 2, kde_data.max() + 2, 200)
ax2.plot(x_range, kde(x_range), color='#e75480', linewidth=2)
ax2.fill_between(x_range, kde(x_range), alpha=0.15, color='#e75480')
ax2.set_title('KDE — Distribution of National Obesity Rates Across Years', fontsize=13)
ax2.set_xlabel('Obesity Rate (%)')
ax2.set_ylabel('Density')

plt.tight_layout()
plt.show()

# --- Plot 2: Obesity rate by state with KDE ---
latest_year = obesity['YearStart'].max()
by_state = obesity[
    (obesity['YearStart'] == latest_year) &
    (obesity['LocationDesc'] != 'National')
].groupby('LocationDesc')['Data_Value'].mean().sort_values(ascending=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 10), gridspec_kw={'width_ratios': [3, 1]})

# Bar chart
ax1.barh(by_state.index, by_state.values, color='#f4a7b9', edgecolor='none')
ax1.axvline(by_state.mean(), color='#c2185b', linewidth=1.5, linestyle='--', label=f'Avg: {by_state.mean():.1f}%')
ax1.set_title(f'Adult Obesity Rate by State ({latest_year})', fontsize=14)
ax1.set_xlabel('Obesity Rate (%)')
ax1.legend()

# KDE alongside bar chart
kde_state = gaussian_kde(by_state.values)
y_range = np.linspace(by_state.min() - 1, by_state.max() + 1, 200)
ax2.plot(kde_state(y_range), y_range, color='#e75480', linewidth=2)
ax2.fill_betweenx(y_range, kde_state(y_range), alpha=0.15, color='#e75480')
ax2.axhline(by_state.mean(), color='#c2185b', linewidth=1.5, linestyle='--')
ax2.set_title('KDE', fontsize=13)
ax2.set_xlabel('Density')
ax2.set_yticklabels([])

plt.tight_layout()
plt.show()

# --- Plot 3: Obesity by income level with KDE ---
income = df[
    (df['Question'] == 'Percent of adults aged 18 years and older who have obesity') &
    (df['StratificationCategory1'] == 'Income') &
    (df['Data_Value_Type'] == 'Value') &
    (df['LocationDesc'] == 'National')
].copy()

income['Data_Value'] = pd.to_numeric(income['Data_Value'], errors='coerce')
income = income.dropna(subset=['Data_Value'])
by_income = income.groupby('Stratification1')['Data_Value'].mean().sort_values()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), gridspec_kw={'width_ratios': [3, 1]})

# Bar chart
ax1.barh(by_income.index, by_income.values, color='#f4a7b9', edgecolor='none')
ax1.axvline(by_income.mean(), color='#c2185b', linewidth=1.5, linestyle='--', label=f'Avg: {by_income.mean():.1f}%')
ax1.set_title('Adult Obesity Rate by Income Level (National)', fontsize=14)
ax1.set_xlabel('Obesity Rate (%)')
ax1.legend()

# KDE alongside bar chart
kde_income = gaussian_kde(by_income.values)
y_range = np.linspace(by_income.min() - 1, by_income.max() + 1, 200)
ax2.plot(kde_income(y_range), y_range, color='#e75480', linewidth=2)
ax2.fill_betweenx(y_range, kde_income(y_range), alpha=0.15, color='#e75480')
ax2.axhline(by_income.mean(), color='#c2185b', linewidth=1.5, linestyle='--')
ax2.set_title('KDE', fontsize=13)
ax2.set_xlabel('Density')
ax2.set_yticklabels([])

plt.tight_layout()
plt.show()

print("Done! 3 charts with KDE curves added.")
```

## Discussion

Paragraph 1 — Main findings & hypothesis

The findings of this analysis support the hypothesis that lower-income adults experience higher rates of obesity compared to higher-income adults. National-level data from the CDC BRFSS dataset revealed a clear inverse relationship between income level and obesity prevalence, with adults earning less than $15,000 annually showing the highest average obesity rates and those earning $50,000 or more showing the lowest. These patterns were consistent across the survey years examined, suggesting that income-based disparities in obesity are not only present but persistent at the national level.

Paragraph 2 — Compare with prior research

These findings align with a substantial body of existing literature linking lower socioeconomic status to higher obesity prevalence. Ogden et al. [cite] similarly documented income-based differences in obesity among U.S. adults using nationally representative data, with stronger disparities observed among women. The dietary cost hypothesis proposed by Drewnowski and Specter [cite] offers a partial explanation, as energy-dense, nutrient-poor foods tend to be more affordable and more accessible in low-income communities. The food environment research by Larson, Story, and Nelson [cite] further supports this, demonstrating that low-income neighborhoods have fewer supermarkets and a higher density of fast-food outlets. The present findings are consistent with these prior studies and reinforce that the income-obesity gradient documented in older literature remains detectable in more recent CDC surveillance data.

Paragraph 3 — Public health implications

The persistence of income-based obesity disparities has meaningful implications for public health policy and intervention design. Programs targeting obesity prevention should prioritize low-income populations and address structural barriers such as food access, neighborhood safety for physical activity, and chronic stress. Evidence-based interventions such as expanding SNAP benefits, subsidizing healthy food retailers in underserved areas, and increasing funding for community-based physical activity programs may help reduce this disparity. These findings also underscore the importance of integrating social determinants of health into chronic disease prevention frameworks, rather than treating obesity as a purely individual behavioral issue.

Paragraph 4 — Limitations

Several limitations of this study should be acknowledged. First, the CDC BRFSS dataset is cross-sectional in nature, meaning that causal relationships between income and obesity cannot be established from these data alone. Second, obesity prevalence values were averaged across multiple survey years for each income group, which may obscure year-to-year variation or changes in the income-obesity gap over time. Third, the income categories used by the CDC are relatively broad, which may mask finer-grained differences within income groups. Finally, the analysis relied on self-reported BMI data, which is subject to measurement error and may systematically underestimate true obesity prevalence. Future research should use longitudinal designs and more granular socioeconomic measures to better characterize the causal pathways linking income to obesity.

## Conclusion

This study examined the relationship between income level and adult obesity prevalence in the United States using nationally representative surveillance data from the CDC BRFSS dataset. The findings support the hypothesis that lower-income adults experience disproportionately higher rates of obesity, with a clear inverse gradient observed across income categories at the national level. These results are consistent with existing literature on socioeconomic determinants of health and reinforce the need for targeted public health interventions that address structural barriers to healthy eating and physical activity among low-income populations. Future research should employ longitudinal designs and incorporate intersecting factors such as race, geography, and gender to more fully characterize the drivers of income-based obesity disparities.
