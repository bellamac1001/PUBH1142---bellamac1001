How do you create a histogram in Seaborn for a variable x in a DataFrame df?

Stacked Histogram with Hue

```python
ax = sns.histplot(data=df, x='x', hue='y', multiple='stack')
```

Heatmap

```python
ax = sns.heatmap(df[['x', 'y']].corr())
```

Box Plot

```python
ax = sns.boxplot(data=df, x='x', y='y')
```

Violin Plot

```python
ax = sns.violinplot(data=df, x='x', y='y')
```

Bar Plot

```python
ax = sns.barplot(data=df, x='x', y='y')
```

Add Title

```python
ax.set_title('My Title')
```

Save Plot

```python
ax.get_figure().savefig('output.png')
```

Set Axis Labels

```python
ax.set(xlabel='X Label', ylabel='Y Label')
```

Change Color Palette

```python
sns.set_palette('Blues')
```

Logarityhmic Y-Axis

```python
ax.set_yscale('log')
```

Regression Line on Scatter Plot

```python
ax = sns.regplot(data=df, x='x', y='y')
```
