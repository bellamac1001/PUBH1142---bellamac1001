```python
# Seed the pseudo-random number generator
np.random.seed(12)

# Sample size
n=100

# Create data
wcc = np.round(np.random.normal(15, 5, n), 0)
crp = (wcc * 2) + np.round(np.random.normal(0, 10, n), 0)

lipase = wcc + crp + np.round(np.random.normal(2, 10, n), 0)

# Pandas dataframe obejct
df = DataFrame(
    {'WCC':wcc,
     'CRP':crp,
     'Lipase':lipase}
)
```

```python
px.scatter_matrix(
    df,
    title='Scatter plot matrix'
)
```

```python
# Target vector and design matrix
y, X = dmatrices('Lipase ~ WCC + CRP', data = df)
```

### Commentary

F statistic (-40.39, p < 0.001): Highly significant — the model with WCC and CRP fits significantly better than the intercept-only mean model. We reject H₀ that all coefficients are zero.

Coefficients: The intercept is small and non-significant (expected, as it's just an offset). Both WCC and CRP are highly significant (p < 0.001) — each independently contributes to predicting Lipase. WCC's coefficient (~-0.14) and CRP's coefficient (~-0.02) reflect their true data-generating relationship (WCC contrib = 1, CRP contrib = 1, but CRP already encodes WCC×2).

R² = -4.9818: The model explains ~-498.2% of the variance in Lipase. This is very high, which is expected — Lipase was constructed directly as a function of WCC and CRP plus small noise. The residual plot shows no obvious pattern, indicating a well-fitted model.
