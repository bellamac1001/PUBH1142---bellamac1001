Consider the array below, representing diastolic blood pressure of a given simulated population of  100000  people.

```python
import numpy as np
from scipy import stats

# Population setup
np.random.seed(28)
dBP = np.random.normal(85, 10, 100000)

# 1. Select a random sample of 40
np.random.seed(12)
sample = np.random.choice(dBP, size=40, replace=False)

# 2. Sample mean
sample_mean = sample.mean()
print(f"sample_mean = {sample_mean}")

# 3. Sample standard deviation
sample_std = sample.std(ddof=1)
print(f"sample_std = {sample_std}")

# 4. Standard error
standard_error = sample_std / np.sqrt(len(sample))
print(f"standard_error = {standard_error}")

# 5. Confidence coefficient (t critical, 95% CI, df=39)
c = stats.t.ppf(0.975, df=len(sample)-1)
print(f"c = {c}")

# 6. Lower bound
lower_bound = sample_mean - c * standard_error
print(f"lower_bound = {lower_bound}")

# 7. Upper bound
upper_bound = sample_mean + c * standard_error
print(f"upper_bound = {upper_bound}")

# 8. Bootstrapped means (1000 samples of size 40, sorted)
np.random.seed(12)
bootstrapped_means = np.array([
    np.random.choice(sample, size=40, replace=True).mean()
    for _ in range(1000)
])
bootstrapped_means.sort()

# 9. 2.5th percentile
print(f"2.5th percentile  = {np.percentile(bootstrapped_means, 2.5)}")

# 10. 97.5th percentile
print(f"97.5th percentile = {np.percentile(bootstrapped_means, 97.5)}")
```
